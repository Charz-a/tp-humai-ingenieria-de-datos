```python
import re

def limpieza_nombre(x):

    con_palo = False
    sin_palo = False
    tacc = False
    
    if 'con palo' in x:
        con_palo = True
    if 'sin palo' in x:
        sin_palo = True
    if 'tacc' in x:
        tacc = True
    
    peso = re.search(r'\d.*', x)
    print(peso.group())
    
            
    return [peso, tacc, con_palo, sin_palo]
    
df_prod[['peso', 'tacc', 'con_palo', 'sin_palo']] = df_prod.nombre.apply(lambda x: limpieza_nombre(x.lower()))

## de momento trabajamos con 500gr y 1kg
## se puede hacer por mas tipos de peso
# df_prod['peso'] = df_prod['nombre'].apply(lambda x: '500 Gr' if '500 Gr' in x else ('1 Kg' if '1 Kg' in x else 'otro'))
```
