```python
def limpieza_nombre(row):

    x = row.nombre.lower()
    con_palo = False
    sin_palo = False
    tacc = False
    
    if 'con palo' in x:
        con_palo = True
    if 'sin palo' in x:
        sin_palo = True
    if 'tacc' in x:
        tacc = True
    
    peso = re.findall(r'\d+ [gr|kg|un].', x)
    # print(peso.group())
    
            
    return peso, tacc, con_palo, sin_palo


df_prod[['peso', 'tacc', 'con_palo', 'sin_palo']] = df_prod.apply(limpieza_nombre, axis=1, result_type ='expand')

df_prod
```
